### Inport neccessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_recall_curve,roc_curve,roc_auc_score

from imblearn.over_sampling import SMOTE
```

### Data preproccessing

```python
df = pd.read_csv('anc dataset.csv')
dummies = pd.get_dummies(df.Presentation, prefix='Presentaion')
dfConcat = pd.concat([df,dummies],axis='columns')
dfFinal = dfConcat.drop(['Presentation'],axis='columns')

X = dfFinal.drop('Complication',axis='columns')
y = dfFinal.Complication
```

### Dataset spliting

```python
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10, stratify=y)
```

### Model fitting

```python
modelLogistic = LogisticRegression(solver='lbfgs', max_iter=1000)
modelSVM = SVC(probability=True)
modelKNN = KNeighborsClassifier(n_neighbors=5)
modelRandomforest = RandomForestClassifier()
#model fitting
modelLogistic.fit(X_train,y_train)
modelSVM.fit(X_train,y_train)
modelKNN.fit(X_train,y_train)
modelRandomforest.fit(X_train,y_train)
```

### Predict probabilities

```python
lr_probs_Logistic = modelLogistic.predict_proba(X_test)
lr_probs_SVM = modelSVM.predict_proba(X_test)
lr_probs_KNN = modelKNN.predict_proba(X_test)
lr_probs_Randomforest = modelRandomforest.predict_proba(X_test)


lr_probs_Logistic = lr_probs_Logistic[:, 1]
lr_probs_SVM = lr_probs_SVM[:, 1]
lr_probs_KNN = lr_probs_KNN[:, 1]
lr_probs_Randomforest = lr_probs_Randomforest[:, 1]

#yhat = model.predict(X_test)
```

### Calculate ROC curve

```python
fpr_logistic, tpr_logistic, _ = roc_curve(y_test, lr_probs_Logistic)
fpr_svm, tpr_svm, _ = roc_curve(y_test, lr_probs_SVM)
fpr_knn, tpr_knn, _ = roc_curve(y_test, lr_probs_KNN)
fpr_randomf, tpr_randomf, _ = roc_curve(y_test, lr_probs_Randomforest)
```

### Calculate ROC AUC

```python
roc_auc_logistic = roc_auc_score(y_test, lr_probs_Logistic)
roc_auc_svm = roc_auc_score(y_test, lr_probs_SVM)
roc_auc_knn = roc_auc_score(y_test, lr_probs_KNN)
roc_auc_randomf = roc_auc_score(y_test, lr_probs_Randomforest)
```

### Ploting ROC curve on imbalanced data

```python
#lw = 2
#complicated = len(y_test[y_test==1]) / len(y_test)
#plt.plot([0, 1], [complicated, complicated], linestyle='--', label='Complicated')
#plt.plot(lr_recall, lr_precision, marker='.', label='Logistic')
#plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')

plt.plot([0, 1], [0, 1], linestyle='--', label='No skill (AUC=0.00)')
plt.plot(fpr_logistic, tpr_logistic, marker='.', color='darkorange', label='Logistic (AUC = %.3f)' % roc_auc_logistic)
plt.plot(fpr_svm, tpr_svm, marker='.', color='blue', label='SVM (AUC = %.3f)' % roc_auc_svm)
plt.plot(fpr_knn, tpr_knn, marker='.', color='red', label='KNN (AUC = %.3f)' % roc_auc_knn)
plt.plot(fpr_randomf, tpr_randomf, marker='.', color='green', label='Random forest (AUC = %.3f)' % roc_auc_randomf)

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve ')
plt.legend(loc="lower right")
plt.show()
```
